Multi inheritance

```python
class mother:
  def __init__(self, fname, lname):
    self.fname = fname
    self.lname = lname

  def getfname(self):
    return self.fname
  def getlname(self):
    return self.lname

class father:
  def __init__(self, age, add):
    self.age = age
    self.add = add

  def getage(self):
    return self.age
  def getadd(self):
    return self.add


class daughter(mother, father):
  def __init__(self, fname, lname, age, add):
    super().__init__(fname, lname)
    super().__init__(age, add)

class son(mother, father):
  def __init__(self,  fname, lname, age, add):
    super().__init__(fname, lname)
    super().__init__(age, add)
```

```python
dau1 = daughter("ab","cd",20,"abcdefg")
print(dau1.lname)
```

```python
class mother1:
  fname = "xyz"
  age = 25
  def __init__(self):
    self.fname = "mother Fname"
    self.lname = "mother Lname"
    self.tmp = "tmp var for mot"

class son1(mother1):
  age = 18
  def __init__(self):
    self.age = 20
    self.add = "Son Add"
    self.tmp1 = "tmp var for son"
    self.fname = "Son fname"
    mother1.__init__(self)
```

Instance var overrides class var with same name

Parent instance var overrides child instance var with same name

Child instance var overrides parent class var

i.e. instance var overrides class var

```python
obj_1 = mother1()
obj_2 = son1()

print(obj_1.fname)
print(obj_1.lname)
newline()
print(obj_2.fname)
print(obj_2.lname)
print(obj_2.age)
print(obj_2.add)
```
