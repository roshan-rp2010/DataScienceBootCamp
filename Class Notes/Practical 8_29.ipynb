Create a Shape class with subclasses like Circle, Rectangle, and Triangle. Implement methods to calculate area and perimeter for each shape.


```python
import math

class Shape:
  def __init__(shape):
    pass

  def Area(shape,area):
    print("The area is: ", round(area,2))

  def Perimeter(shape,perm):
    print("The perimeter is: ", round(perm,2))
```

```python
class Circle(Shape):
  def __init__(circle,radius):
    if radius > 0:
      circle.radius = radius
    else:
      circle.radius = 0

  def printArea(circle):
    circle.Area(math.pi * circle.radius * circle.radius)

  def printPerimeter(circle):
    circle.Perimeter(2 * math.pi * circle.radius)

class Rectangle(Shape):
  def __init__(rectangle,length,width):
    if length > 0:
      rectangle.length = length
    else:
      rectangle.length = 0
    if width > 0:
      rectangle.width = width
    else:
      rectangle.width = 0

  def printArea(rectangle):
    rectangle.Area(rectangle.length * rectangle.width)

  def printPerimeter(rectangle):
    rectangle.Perimeter(2 * (rectangle.length * rectangle.width))

class Triangle(Shape):
  def __init__(triangle,side1,side2,side3):
    if side1 > 0:
      triangle.side1 = side1
    else:
      triangle.side1 = 0
    if side2 > 0:
      triangle.side2 = side2
    else:
      triangle.side2 = 0
    if side3 > 0:
      triangle.side3 = side3
    else:
      triangle.side3 = 0

  def printArea(triangle):
    s = (triangle.side1 + triangle.side2 + triangle.side3)/2
    area = math.sqrt(s * (s - triangle.side1) * (s - triangle.side2) * (s - triangle.side3))
    triangle.Area(area)

  def printPerimeter(triangle):
    s = (triangle.side1 + triangle.side2 + triangle.side3)
    triangle.Perimeter(s)
```

```python
cir1 = Circle(5)
cir1.printArea()
cir1.printPerimeter()
print()
rec1 = Rectangle(4,1)
rec1.printArea()
rec1.printPerimeter()
print()
tri1 = Triangle(2,3,4)
tri1.printArea()
tri1.printPerimeter()
```

Online Shopping Cart: Develop an Item class and a ShoppingCart class. The ShoppingCart class should allow adding items, removing items, calculating the total cost, and applying discounts.

```python
class Item:
  def __init__(this,name,cost):
    if isinstance(name, str):
      this.name = name
    else:
      this.name = ""
    if isinstance(cost, float) or isinstance(cost, int):
      this.cost = cost
    else:
      this.cost = 0.0

  def getName(this):
    return this.name
  def getCost(this):
    return this.cost
```

```python
class ShoppingCart(Item):
  def __init__(this):
    this.icart = []

  def addItem(this, item: Item):
    this.icart.append(item)

  def removeItem(this, item: Item):
    found = False
    for i in this.icart:
      if i.name == item.name and i.cost == item.cost:
        this.icart.remove(i)
        found = True
        break
    if not found:
      print("Given item doesn't exist in the cart!\n")

  def showCart(this):
    for i in this.icart:
      print((i.getName()).ljust(40), end = "")
      print(str(i.getCost()))

  def totalCost(this):
    tmpCost = 0
    for i in this.icart:
      tmpCost = tmpCost + i.getCost()
    print("Your total cost = " + str(tmpCost))

  def applyDiscount(this, dis_per: float):
    tmpCost = 0
    if dis_per < 100 and dis_per > 0:
      for i in this.icart:
        tmpCost = tmpCost + i.getCost()
      print("Your total cost after discount = " + str(tmpCost - tmpCost * (dis_per/100)))
    else:
      print("Discount percentage must be less than 100 and greater than 0!")
```

```python
itm1 = Item("Rice", 30)
itm2 = Item("Beef", 50)
itm3 = Item("Vegetable", 20)

mycart = ShoppingCart()
print(mycart.showCart())
print()

mycart.addItem(itm1)
mycart.addItem(itm2)
mycart.addItem(itm3)
mycart.addItem(Item("garbage bag", 5))
mycart.showCart()
print()
mycart.removeItem(Item("garbage bag", 5))
mycart.showCart()
print()

mycart.totalCost()
print()

mycart.applyDiscount(20)
```

Social Media System: Create classes for User, Post, and Comment. Users can create posts, like posts, and comment on posts. Implement methods to display posts with their interactions.

```python
class Comment:
  def __init__(this,comment: str = ""):
    this.comment = comment

# private method
# returns comment in string format
  def _getComment(this):
    return this.comment

class Post(Comment):
# initializes post with post or empty string, likes with 0 likes and empty comment list
  def __init__(this, post: str = ""):
    this.post = post
    this.likes = 0
    this.comment = []

# prints the post itself
  def getPost(this):
    print("This post is : ", this.post)

# prints total number of likes the post has
  def getLikes(this):
    print("This post has : ", this.likes, "likes")

# prints all the comments of that post
  def getComments(this):
    print("This post has the following comments:")

    for i in this.comment:
      print("-   ", i._getComment())

# private method
# adds 1 like to the post. 1 is passed as param from user class
  def _addLikes(this, num: int):
    if num > 0:
      this.likes += num
    else:
      print("Likes must be more than 0!")

# private method
# adds comment in the post
  def _addComment(this, com: Comment):
    this.comment.append(com)


class User(Post):
  def __init__(this):
    this.post = Post()

# allows users to create post and maybe abuse it to update post
  def createPost(this, post: Post):
    this.post = post

# allows users to like other's post
  def likePost(this, post: Post):
    post._addLikes(1)

# allows users to comment on other's posts
  def commentOnPost(this, post: Post, com: Comment):
    post._addComment(com)
```

```python
myuser1 = User()

user1_post = Post("My First Post!")
myuser1.createPost(user1_post)

myuser2 = User()
myuser2.likePost(user1_post)
myuser2_comment = Comment("Congratulation on your first post!")
myuser2.commentOnPost(user1_post,myuser2_comment)

myuser3 = User()
myuser3.likePost(user1_post)
myuser3_comment = Comment("Nice Post!")
myuser3.commentOnPost(user1_post,myuser3_comment)

myuser4 = User()
myuser4.likePost(user1_post)
myuser4_comment = Comment("Welcome to social media!")
myuser4.commentOnPost(user1_post,myuser4_comment)

myuser5 = User()
myuser5_comment = Comment("What a stupid post!")
myuser5.commentOnPost(user1_post,myuser5_comment)

myuser6 = User()
myuser6.likePost(user1_post)

myuser7 = User()
myuser7.likePost(user1_post)

user1_post.getPost()
print()
user1_post.getLikes()
print()
user1_post.getComments()
```
