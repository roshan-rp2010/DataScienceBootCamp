```python
import pandas as pd
import numpy as np

path = "/content/demo_data.xlsx - Sheet1.csv"
df = pd.read_csv(path)
```

```python
df_updated = df.dropna()
```

```python
dictionary_of_mean_wage = {}

for k in df_updated.Education.value_counts().keys():
  dictionary_of_mean_wage[k] = df_updated.Wage[df["Education"] == k].mean()

print(dictionary_of_mean_wage)

k = list(dictionary_of_mean_wage.keys())
v = list(dictionary_of_mean_wage.values())

print(k[v.index(max(v))])
```
