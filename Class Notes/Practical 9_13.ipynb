# Pascal's triangle pattern

```python
ptlist = [[1],[1,1]]
ptlist_instr = []
tot_line = int(input("Total lines of pascal's triangle (must be positive int): "))
for i in range(tot_line+1):
  nextline = []
  if i > 1:
    nextline.append(1)
    for j in range(1,i):
      nextline.append(ptlist[i-1][j-1] + ptlist[i-1][j])
    nextline.append(1)
    ptlist.append(nextline)
  ptlist_instr.append(" ".join(str(i) for i in ptlist[i]))

mylen = len(ptlist_instr[-1])

for i in ptlist_instr:
  print(i.center(mylen," "))
```
