```python
class person: # parent class
  def __init__(self, fname, mname, lname):
    pass
  def print_details(self):
    print(self.fname, self.mname, self.lname)

class student(person): # child class
  def __init__(self, fname, mname, lname):
    pass
  def print_details(self):
    print(self.fname, self.mname, self.lname)
```

```python
class Car:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Drive!")

class Boat:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Sail!")

class Plane:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Fly!")

car1 = Car("Ford", "Mustang")       #Create a Car class
boat1 = Boat("Ibiza", "Touring 20") #Create a Boat class
plane1 = Plane("Boeing", "747")     #Create a Plane class

# overloading move with different x, overloading x with different variables
for x in (car1, boat1, plane1):
  x.move()
```

```python
class vehicle:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print('move!!')

# inherits from vehicles
class Car(vehicle):
  # child class methods overrides parent class methods with same calling name
  def move(self):
    print("Drive!")

class Boat(vehicle):
  def move(self):
    print("Sail!")

class Plane(vehicle):
  def move(self):
    print("Fly!")

car1 = Car("Ford", "Mustang")       #Create a Car class
boat1 = Boat("Ibiza", "Touring 20") #Create a Boat class
plane1 = Plane("Boeing", "747")     #Create a Plane class

# polymorphism
for i in (car1,boat1,plane1):
  print(i.brand)
  print(i.model)
  i.move()
```

```python
# Inheritance, Polymorphism, Overriding and Encaptulation
class animal:
  def __init__ (self, living_area, leg_count, hair_colour):
    self.living_area = living_area
    self.leg_count = leg_count
    self.hair_colour = hair_colour

  def print_det(self):
    print("This is animal class!")

class dog(animal):
  def print_det(self):
    print("This dog lives on", self.living_area, "has", self.leg_count, "legs", "and", self.hair_colour, "coloured hair!")

class parrot(animal):
  def print_det(self):
    print("This parrot flies in", self.living_area, "has", self.leg_count, "legs", "and", self.hair_colour, "coloured hair!")


#Experiment
class fish(animal):
  def __init__ (self, living_area):
    # can use only one of the two below lines, first assigns empty values to leg and hair variables
    super().__init__(living_area, leg_count="",hair_colour="")
    #self.living_area = living_area

  def print_det(self):
    print("This fish lives in", self.living_area, "has no legs and no hair!")


# main

animal1 = animal("Animal_living_area","Animal_legs","Animal_hair")
dog1 = dog("land",4,"grey")
parrot1 = parrot("air",2,"green")
fish1 = fish("water")

for i in (animal1, dog1, parrot1, fish1):
  print(i.living_area)
  if isinstance(i,fish) == False:
    print(i.leg_count)
    print(i.hair_colour)

  i.print_det()
  newline()
```
