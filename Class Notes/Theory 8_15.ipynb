```python
# hourly glass matrix

for i in range(10):
  for j in range(0, i):
    print("", end = " ")
  for k in range(10, i, -1):
    print("*", end = " ")
  print("\r")

for i in range(1,10):
  for j in range(1, 10 - i):
    print(" ", end = "")
  for k in range(0, i + 1):
    print("*", end = " ")
  print("\r")
```

```python
# hollow hourly glass matrix

for i in range(10):
  for j in range(0, i):
    print("", end = " ")
  for k in range(10, i, -1):
    if i == 0 or k == 10 or k-i == 1:
      print("*", end = "")
    print(" ", end = " ")
  print("\r")

for i in range(1, 10):
  for j in range(1, 10 - i):
    print(" ", end = "")
  for k in range(0, i + 1):
    if k == 0 or i == 9 or k == i:
      print("*", end = "")
    print(" ", end = " ")
  print("\r")
```
