```python
fruits_lst = ["apple","banana","cherry","green apple","red apple","pineapple","watermelon"]
lst_com2 = [i for i in fruits_lst]
print(lst_com2)
```

```python
fruits_lst = ["apple","banana","cherry","green apple","red apple","pineapple","watermelon"]
lst_com3 = [i for i in fruits_lst if "apple" not in i]
print(lst_com3)
```

```python
lst = [chr(i) for i in range(97,104)]
lst2 = [i.upper() if i>"c" else i.lower() for i in lst]
lst2
```

```python
lst = [chr(i) for i in range(97,104)]
lst2 = [newlst.upper() if newlst>"d" else i.lower() for i in lst for newlst in i if ord(i)%2 == 0]
lst2
```

```python
lst = [chr(i) for i in range(97,104)]
lst2 = [j.upper() for newlst in lst if ord(newlst) % 2 == 0 for j in newlst if j > "d"]
lst2
```

```python
lst_mtx = [[i for i in range(3)] for i in range(4)]
lst_mtx
```

```python
ndim2array = [[10,20,30],
              [40,50,60],
              [70,80,90]]

lst1 = [[i[j] for i in ndim2array] for j in range(len(ndim2array))]
lst1
```
