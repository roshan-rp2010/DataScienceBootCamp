Bank Account: Create a BankAccount class with methods to deposit, withdraw, and get the current balance. Ensure that the balance can't go negative.

```python
class BankAccount:
  def __init__(bank, acc_bal = 0):
    if acc_bal < 0:
      print("Cannot create account with negative balance!")
      bank.acc_bal = 0
    else:
      bank.acc_bal = acc_bal

  def getBalance(bank):
    return bank.acc_bal

  def deposit(bank, tmp_add):
    if tmp_add > 0:
      bank.acc_bal = bank.acc_bal + tmp_add
    else:
      print("Cannot deposit negative balance!")

  def withdraw(bank, tmp_sub):
    if tmp_sub < bank.acc_bal:
      bank.acc_bal = bank.acc_bal - tmp_sub
    else:
      print("Cannot withdraw more than current balance!")
```

```python
newacc = BankAccount(-100)
print(newacc.getBalance())

newacc.deposit(20)
print(newacc.getBalance())

newacc.withdraw(20)
print(newacc.getBalance())

newacc.withdraw(100)
print(newacc.getBalance())
```

Library Catalog: Implement a Library class with methods to add books, lend books, and return books. Keep track of available and borrowed books.

```python
class Library:
  def __init__(lib, books = []):
    if type(books) == list:
      lib.books = books
    else:
      lib.books = []
    lib.borbooks = []

  def get_available_books(lib):
    return lib.books

  def get_borrowed_books(lib):
    return lib.borbooks

  def add_Books(lib, addbooks):
    if type(addbooks) == list:
      lib.books.extend(addbooks)
    else:
      print("Books names must be in list format!")

  def lend_Books(lib, lendbooks):
    for i in lendbooks:
      if i in lib.books:
        lib.borbooks.extend(i)
        lib.books.remove(i)
      else:
        print("Book \"", i, "\"", "isn't available to lend!")
```

```python
mylib = Library(["A","B","C","D"])
print("Available books: ", mylib.get_available_books())
newline()

mylib.add_Books("E")
print("Available books: ", mylib.get_available_books())
newline()

mylib.add_Books(["F","G"])
print("Available books: ", mylib.get_available_books())
newline()

mylib.lend_Books("A")
print("Available books: ", mylib.get_available_books())
print("Lend books:      ", mylib.get_borrowed_books())
newline()

mylib.lend_Books(["B","C"])
print("Available books: ", mylib.get_available_books())
print("Lend books:      ", mylib.get_borrowed_books())

newline()
mylib.lend_Books(["A","D"])
print("Available books: ", mylib.get_available_books())
print("Lend books:      ", mylib.get_borrowed_books())
```
