```python
class childclass(person):
  pass
```

```python
obj2 = childclass("Anu", "Edlab")
obj2.printName()
```

```python
class childclass(person):
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printName(self):
    print("class childclass(person):", self.firstname, " ", self.lastname)
```

```python
# parent class
class vehicle:
  def __init__(car, make, model):
    car.make = make
    car.model = model

  def printName(car):
    print(car.make, " ", car.model)

# child class, inherted from parent: vehicle
class kidsVehicle(vehicle):
  def __init__(car, make, model):
    car.make = make
    car.model = model
```

```python
# parent class
class vehicle:
  def __init__(car, make, model):
    car.make = make
    car.model = model

  def printName(car):
    print(car.make, " ", car.model)

# child class, inherted from parent: vehicle
class kidsVehicle(vehicle):
  def __init__(car, make, model):
    super().__init__(make, model)
    # having the following replaces any provided argument for those variables
    car.model = "Universal 2020"
```
