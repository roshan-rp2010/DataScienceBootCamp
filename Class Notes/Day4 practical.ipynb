```python
light = input("Light Color: ").lower()

if light == "green":
  print("Car is allowed to go")
elif light == "yellow":
  print("Car has to wait")
elif light == "red":
  print("Car has to stop")
else:
  print("unrecognized signal")
```

```python
marks = int(input("Subject Mark: "))

if(marks < 0 or marks > 100):
  print("error: mark should be between o and 100 only")
else:
  if marks < 50:
    print("Fail")
  elif marks >= 50 and marks < 60:
    print("Good")
  elif marks >= 60 and marks < 80:
    print("Very Good")
  else:
    print("Outstanding")

  print("Marks obtained = ", marks)
```
