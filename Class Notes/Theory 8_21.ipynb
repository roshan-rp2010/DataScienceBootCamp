```python
from IPython.core.debugger import make_arrow
class cars:
  def __init__ (car, make, model, year, vin):
    car.make = make
    car.model = model
    car.year = year
    car.vin = vin

  def print_info(car):
    print("The make of car is  :  ", car.make)
    print("The model of car is :  ", car.model)
    print("The year of car is  :  ", car.year)
    print("The vin of car is   :  ", car.vin)

# get and set make
  def getmake(car):
    return car.make

  def setmake(car, make):
    car.make = make

# get and set model
  def getmodel(car):
    return car.model

  def setmodel(car, model):
    car.model = model

# get and set year
  def getyear(car):
    return car.year

  def setyear(car, year):
    car.year = year

# get and set vin
  def getvin(car):
    return car.vin

  def setvin(car, vin):
    car.vin = vin
```

```python
new_car1 = cars("Ford", "Ram1500", 2020, 3498537532)
new_car1.print_info()

print("\nNew Car details:\n")

new_car1.setmake("Chevy")
print("The make of car is  :  ", new_car1.getmake())

new_car1.setmodel("Impala")
print("The model of car is :  ", new_car1.getmodel())

new_car1.setyear(2022)
print("The year of car is  :  ", new_car1.getyear())

new_car1.setvin(197342928)
print("The vin of car is   :  ", new_car1.getvin())
```

```python
# great great grandparent
class Animals:
  pass

# great grandparent
# inherits from great great grandparent: Animals
class Mammals(Animals):
  pass
class Amphebians(Animals):
  pass
class Insects(Animals):
  pass
class Aquatic(Animals):
  pass
class Birds(Animals):
  pass

# grandparent
# inherits from great grandparent: Mammals
class legged2(Mammals):
  pass
class legged4(Mammals):
  pass

# grandparent
# inherits from great grandparent: Aquatic
class no_legs(Aquatic):
  pass
class lost_of_legs(Aquatic):
  pass

# parent
# inherits from grandparent: legged2
class homo_sapiens(legged2):
  pass
class kangaroo(legged2):
  pass

# parent
# inherits from grandparent: legged4
class cow(legged4):
  pass
class goat(legged4):
  pass

# parent
# inherits from grandparent: no_legs
class fish(no_legs):
  pass

# parent
# inherits from grandparent: lots_of_legs
class octopus(lost_of_legs):
  pass

# child
# inherits from parent: homo_sapiens
class man(homo_sapiens):
  pass
class woman(homo_sapiens):
  pass
```
