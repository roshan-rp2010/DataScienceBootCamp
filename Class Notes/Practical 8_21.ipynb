1) Write a Python Function that accepts a hyphen-separated sequence of words as input and prints the words in a hyphen-separated sequence after sorting them alphabetically.

Sample Items : green-red-yellow-black-white

Expected Result : black-green-red-white-yellow

```python
# sorts the words seperated by hyphen '-' in alphabetical order

def sort_the_strings(seq):
  list1 = seq.split("-")
  list1.sort()
  print(*list1, sep = "-")

  # alternative method to print the list
  #print("-".join(list1))


#main

mystr = input("Enter the string seperated with hyphen: ")
newline()
sort_the_strings(mystr)
```

2) Write a function that accepts a string and count consonant and vowels in that string

```python
# counts total vowels and consonants in the given string

def count_vowels_and_consonants(mystr1):
  mystr1 = mystr1.lower()
  vowel_list = ["a", "e", "i", "o", "u"]
  totvow = 0
  totcon = 0

  for i in mystr1:
    if i >= "a" and i <= "z":
      if i in vowel_list:
        totvow += 1
      else:
        totcon += 1

  print("Total vowels in the string = ", totvow)
  print("Total consonants in the string = ", totcon)


#main

mystr = input("Enter the string: ")
newline()
count_vowels_and_consonants(mystr)
```

3) Write a Python function to check whether a number is "Perfect" or not.
According to Wikipedia : In number theory, a perfect number is a positive integer that is equal to the sum of its proper positive divisors, that is, the sum of its positive divisors excluding the number itself (also known as its aliquot sum). Equivalently, a perfect number is a number that is half the sum of all of its positive divisors (including itself).

Example : The first perfect number is 6, because 1, 2, and 3 are its proper positive divisors, and 1 + 2 + 3 = 6. Equivalently, the number 6 is equal to half the sum of all its positive divisors: ( 1 + 2 + 3 + 6 ) / 2 = 6. The next perfect number is 28 = 1 + 2 + 4 + 7 + 14. This is followed by the perfect numbers 496 and 8128.

```python
# checks if the number is perfect or imperfect

def perfect_imperfect(myint1):
  mylist1 = []
  for i in range(1,myint1):
    if myint1%i == 0:
      mylist1.append(i)
  netsum = 0
  for i in mylist1:
    netsum = netsum + i

  if netsum == myint1 or (netsum + myint1)/2 == myint1:
    print("The number", myint1, " is a perfect number!")
  else:
    print("The number", myint1, " is not a perfect number!")


#main

myint = int(input("Enter a number to check it's perfectness: "))
newline()
perfect_imperfect(myint)
```

4) Write a program to print twin primes less than 1000. If two consecutive odd numbers are both prime then they are known as twin primes

```python
# prints a list of twin primes between 2 and 1000. Twin primes don't include even numbers. i.e. 2

def twin_primes():
  mylist1 = []
  for i in range(3,1000):
    for j in range(2,i):
      if i%j == 0:
        break
      if j == i-1:
        mylist1.append(i)

  for i in range(0,len(mylist1)-2):
    if mylist1[i]+2 == mylist1[i+1]:
      print(mylist1[i],",",mylist1[i+1])


#main

twin_primes()
```

5) Write a program to find out the prime factorization of a number.

Example: prime factorization of 56 is 2 × 2 × 2 × 7

```python
# prints prime factors of given number

def prime_fact(myint1):
  i = 2
  mylist1 = []

  while i*i <= myint1:
    if myint1%i == 0:
      mylist1.append(i)
      myint1 = myint1//i
    else:
      i += 1

  if myint1 > 1:
    mylist1.append(myint1)

  print(mylist1)


#main

myint = int(input("Enter a number to print it's prime factors: "))
newline()
prime_fact(myint)
```
