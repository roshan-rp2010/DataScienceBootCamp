#**Regression**

Machine learning regression is a technique for investigating the relationship between independent variables or features and a dependent variable or outcome. It's used as a method for predictive modelling in machine learning, in which an algorithm is used to predict continuous outcomes.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
path = "/content/homeprices.csv"
df = pd.read_csv(path)
df
```

```python
import matplotlib.pyplot as plt

plt.scatter(df.area, df.price, color= "red", marker= "*")
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("House Price Analysis")
plt.show()
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()
df_new = df.drop("price", axis= "columns")
price = df.price
reg.fit(df_new, price)
```

```python
reg.predict([[3300]])[0]
```

```python
path = "/content/Salary_Data.csv"
df = pd.read_csv(path)
df
```

```python
plt.scatter(df.YearsExperience, df.Salary, color= "red", marker= "*")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary Analysis")
plt.show()
```

```python
X = df.iloc[:, 0:1].values
y = df.iloc[:, 1:2].values
X, y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 1/3, random_state= 1)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.predict([[5.5]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
(9158.13919873 * 5.5) + 26137.2400142
```
