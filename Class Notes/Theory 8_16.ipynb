```python
# fibonacci series

total = int(input("total series numbers: "))

fibb = [0, 1, 1]

for i in range(total-3):
  fibb.append(fibb[-1] + fibb[-2])

print(fibb)
```

```python
# grade distinction

exam_marks = {"Sameer":89, "Sabina":87, "Inder":37, "Keshav":67, "Rohit":77, "Suman":55}

for keys,values in exam_marks.items():
  if values > 80:
    print(keys, values, "Dist")
  elif values > 70:
    print(keys, values, "First")
  elif values > 60:
    print(keys, values, "Second")
  elif values > 50:
    print(keys, values, "Third")
  else:
    print(keys, values, "Fail")
```
