```python
list1 = ["Apple","Banana","Mango","Peach"]
print("Negative index negative slicing")
print(list1[-1::-1])
print("Positive index positive slicing")
print(list1[::1])
print("Positive index negative slicing")
print(list1[::-1])
print("Negative index positive slicing")
print(list1[-len(list1)::1])
```
