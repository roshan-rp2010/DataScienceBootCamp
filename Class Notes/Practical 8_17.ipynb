```python
my_num = int(input("Number: "))
fact = 1

for i in range(1, my_num + 1):
  fact = fact*i

print("Factorial of", my_num, "is:", fact)
```

```python
# fibonacci series already done in theory yesterday

tot = int(input("Total fibonacci numbers to be printed: "))

fib = [0,1]

for i in range(tot-2):
  fib.append(fib[-1] + fib[-2])

print(fib[:tot:])
```
