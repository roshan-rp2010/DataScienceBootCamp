```python
print("string.upper() \nconverts string to upper case")
print(str1.upper(),"\n")
print("string.lower() \nconverts string to lower case")
print(str1.lower(),"\n")
print("string.casefold() \nconverts string to lower case but much more aggresively, example: 11hELlO45")
print("11hELlO45".casefold(),"\n")
print("string.center(width,fill_char) \nfills the additional width with fill char")
print(str1.center(21,"-"),"\n")
print("string.count() \nfinds number of substring occourance in the given string, case sensative")
print(str1.count(str1),"\n")


```
