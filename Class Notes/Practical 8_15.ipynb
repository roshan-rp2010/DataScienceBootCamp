```python
# for loop practice

a_list = [1,2,3,4,5,11,1,2,6,1,8]
b_list = []

for i in a_list:
  if i not in b_list:
    b_list.append(i)

print(b_list)
```

```python
# for loop practice
prime_list = []
i = 2
j = 2

for i in range(2,100):
  for j in range(2,i+1):
    if i%j == 0:
      break
  if i == j:
    prime_list.append(i)

print(prime_list)
```

```python
a_list = [1,2,3,4,5,6,7,8,9]

#first method
print("[", end = "")
for i in range(1,len(a_list)+1):
  print(a_list[-1*i], end = "")
  if i != len(a_list):
    print(", ", end = "")
print("]\n")

# second method
rev_list = []

for i in range(len(a_list)):
  rev_list.append(a_list.pop())

print(rev_list)
```

```python
num_1 = int(input("first number: "))
num_2 = int(input("second number: "))

gcd = 1

for i in range(num_1,0,-1):
  if num_1%i == 0:
    if num_2%i == 0:
      gcd = i
      break

print("gcd: ", gcd)
```
