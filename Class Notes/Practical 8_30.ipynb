Vehicle Rental System: Build classes for Vehicle, Car, and Bike. Implement a rental system where customers can rent vehicles and the system can track availability.

```python
class Car:
  def __init__(this, make: str = "", model: str = ""):
    this.make = make
    this.model = model

  def getMake(this):
    return this.make

  def getModel(this):
    return this.model


class Bike:
  def __init__(this, make: str = "", model: str = ""):
    this.make = make
    this.model = model

  def getMake(this):
    return this.make

  def getModel(this):
    return this.model


class Vehicles(Car, Bike):
  def __init__(this):
    this.Vlist = []

  def addVehicle(this, car: Car = None, bike: Bike = None):
    if car is not None:
      this.Vlist.append(car)
    elif bike is not None:
      this.Vlist.append(bike)
    else:
      print("No vehicle info received to add into the inventory!")

  def getAllVehicles(this):
    for i in this.Vlist:
      print((i.getMake()).ljust(15), end = "")
      print(i.getModel())

  def isAvailable(this, car: Car = None, bike: Bike = None):
    if car is not None:
      tmpbool = False
      for i in this.Vlist:
        if i.getMake() == car.getMake() and i.getModel() == car.getModel():
          print(car.getMake(), car.getModel(), "is available!")
          tmpbool = True
          break
      if not tmpbool:
        print(car.getMake(), car.getModel(), "is NOT available!")
    elif bike is not None:
      tmpbool = False
      for i in this.Vlist:
        if i.getMake() == bike.getMake() and i.getModel() == bike.getModel():
          print(bike.getMake(), bike.getModel(), "is available!")
          tmpbool = True
          break
      if not tmpbool:
        print(bike.getMake(), bike.getModel(), "is NOT available!")
    else:
      print("No vehicle info received to check availability!")

  def rentVehicle(this, car: Car = None, bike: Bike = None):
    if car is not None:
      tmpbool = False
      for i in this.Vlist:
        if i.getMake() == car.getMake() and i.getModel() == car.getModel():
          this.Vlist.remove(i)
          print(car.getMake(), car.getModel(), "is now rented!")
          tmpbool = True
          break
      if not tmpbool:
        print(car.getMake(), car.getModel(), "is NOT available to rent!")
    elif bike is not None:
      tmpbool = False
      for i in this.Vlist:
        if i.getMake() == bike.getMake() and i.getModel() == bike.getModel():
          this.Vlist.remove(i)
          print(bike.getMake(), bike.getModel(), "is now rented!")
          tmpbool = True
          break
      if not tmpbool:
        print(bike.getMake(), bike.getModel(), "is NOT available to rent!")
    else:
      print("No vehicle info received to rent!")
```

```python
car1 = Car("Toyota","Avalon")
car2 = Car("Ford","F150")
bike1 = Bike("Yamaha","r7")
bike2 = Bike("Suzuki","GSX1300")
bike3 = Bike("Ninja","H2R")

newcar = Car("Koenigsegg","Jesko")

vehicle = Vehicles()
vehicle.addVehicle(car1)
vehicle.addVehicle(car2)
vehicle.addVehicle(bike1)
vehicle.addVehicle(bike2)
vehicle.addVehicle(bike3)

vehicle.getAllVehicles()
print()

vehicle.isAvailable(car1)
vehicle.isAvailable(bike3)
vehicle.isAvailable(newcar)
print()

vehicle.rentVehicle(bike2)
vehicle.rentVehicle(newcar)
print()

vehicle.getAllVehicles()
```

Online Quiz Application: Develop classes for Question, Quiz, and User. Users can take quizzes, answer questions, and see their scores at the end.

```python
import random
```

```python
class Question:
  def __init__(this, que:str, ans:str):
    if not que.isspace() and not ans.isspace() and que != "" and ans != "":
      this.que = que
      this.ans = ans.lower()
    else:
      raise Exception("Neither question nor answers can be empty!")

  def _getQuestion(this):
    return this.que

  def _getAnswer(this):
    return this.ans

class Quiz(Question):
  def __init__(this):
    this.qset = []

  def addQuestion(this, que: Question):
    this.qset.append(que)

  def printQuestions(this):
    for i, v in enumerate(this.qset):
      print(i+1,". ",v._getQuestion())

  def _takeQuiz(this):
    if len(this.qset) == 0:
      print("Sorry, no questions in Quiz!")
    else:
      # generates a list with length same as this.qset with random numbers between 0 and this.qset
      qorder = random.sample(range(0,len(this.qset)),len(this.qset))
      score = 0
      for i in qorder:
        print("Question: ", this.qset[i]._getQuestion())
        tmpans = input("Answer: ").lower()
        print()
        if tmpans == this.qset[i]._getAnswer():
          score += 1

      print("\nYour total score is : ", score)

  def _answerQuestion(this):
    tryagain = True
    while tryagain:
      this.printQuestions()

      print("\nPlease choose the question you'd like to attempt!\n")

      qnum = -1
      while True:
        try:
          qnum = int(input("Question number: "))
          if qnum > len(this.qset) or qnum < 1:
            raise Exception()
          break
        except Exception as E:
          print("Question number must be in number format and from available question numbers!\n")

      print("\nQuestion: ", this.qset[qnum-1]._getQuestion())
      tmpans = input("Answer: ").lower()
      if tmpans == this.qset[qnum-1]._getAnswer():
        print("\nYour answer is correct!")
      else:
        print("\nYour answer is incorrect!")

      while True:
        tmp_str = (input("\nWould you like to try again? (Y/N)")).upper()
        if tmp_str != "Y" and tmp_str != "N":
          print("Response must be letters Y or N only!")
          continue
        break

      if tmp_str == "Y":
        tryagain = True
      else:
        tryagain = False

      print()

class User(Quiz):
  def __init__(this):
    pass

  def takeQuiz(this, quiz: Quiz):
    quiz._takeQuiz()

  def answerQuestion(this, quiz: Quiz):
    quiz._answerQuestion()
```

School Management System: Design classes for School, Student, and Teacher. Students can enroll in courses, teachers can assign grades, and the school can generate reports.

```python
import numpy as np
import pandas as pd
```

```python
class Student:
  def __init__(this, name: str = "", grade: int = 0):
    this.name = name
    this.grade = grade

  def getName(this):
    return this.name

  def getGrade(this):
    return this.grade


class Teacher(Student):
  def __init__(this, name: str = "", course: str = ""):
    this.name = name
    this.course = course

  def getName(this):
    return this.name

  def getCourse(this):
    return this.course

  def assignGrades(this, student: Student, grade: int):
    student.grade = grade


class School(Teacher, Student):
  def __init__(this):
    this.courses = []
    # later becomes 2d list
    this.studlist = []

  def addCourse(this, course: str = ""):
    this.courses.extend([course.lower()])
    this.studlist.append(["name"])
    this.studlist.append(["grade"])

  def enrollInCourse(this, student: Student, course: str = ""):
    if course.lower() in this.courses:
      ind = this.courses.index(course.lower())
      this.studlist[ind * 2].extend([student.getName()])
      this.studlist[(ind * 2) + 1].extend([student.getGrade()])
    else:
      print("Course doesn't exist!")

  def generateReports(this, course = None):
    newlist = {"student_name":"grade"}
    if course == None:
      for i, v in enumerate(this.studlist[::2]):
        for j, k in enumerate(v[1::]):
          if j in newlist:
            newlist[k] += this.studlist[(i*2)+1][j+1]
          else:
            newlist[k] = this.studlist[(i*2)+1][j+1]
    else:
      if course.lower() in this.courses:
        ind = this.courses.index(course.lower())
        nameind = ind * 2
        gradind = (ind * 2) + 1
        for i, v in enumerate(this.studlist[nameind]):
          newlist[v] = this.studlist[gradind][i]
      else:
        print("Course doesn't exist for grade of students!")

    totgrade = pd.DataFrame.from_dict(newlist, orient = "index", columns = "Grade:")
    print(totgrade)

  def printStudentsList(this):
    newlist = this.studlist[::2]
    studlst = pd.DataFrame(newlist)
    studlst = studlst.transpose()
    studlst = studlst.drop(0)
    studlst.columns = [i.title() for i in this.courses]
    newlst = studlst.replace(np.nan, '', regex=True)
    print(newlst)
```

```python
stud1 = Student("Alpha",0)
stud2 = Student("Beta",80)
stud3 = Student("Gamma",100)
stud4 = Student("Delta",20)
stud5 = Student("Row",50)
stud6 = Student("Phi",46)
course1 = "Computer science"
course2 = "Math"
course3 = "Chemistry"
course4 = "Biology"
course5 = "Physics"

school = School()

school.addCourse(course1)
school.addCourse(course2)
school.addCourse(course3)
school.addCourse(course4)
school.addCourse(course5)

school.enrollInCourse(stud1,course1)
school.enrollInCourse(stud2,course1)
school.enrollInCourse(stud3,course1)
school.enrollInCourse(stud4,course2)
school.enrollInCourse(stud5,course2)
school.enrollInCourse(stud6,course3)

school.printStudentsList()
print()
school.generateReports()
print()
school.generateReports(course1)
```
