```python
class InvalidAngle(Exception):
  "Raise when the input is less than 0 and greater than 360, non-inclusive"
  pass

try:
  input_num = int(input("Your compass angle:"))
  if input_num < 0 or input_num > 360:
    raise InvalidAngle
  else:
    print("Your heading is:",input_num)

except InvalidAngle as ia:
  print("exception as Invalid Angle:")
```
