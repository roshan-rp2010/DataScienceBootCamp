```python
# need to ask user for a number
# ask user for how many rows needed

# number of rows is basically where iteration stops
# csv should be 3 columns
# 1st column is like 5,5,10,15,25,40,... [i] + [i+1]
# 2nd column is like 5,25,625,... square
# 3rd column is like 5,125,... cube
```

```python
import os
import os.path
import pandas as pd

totrow = int(input("Enter number of rows: "))
num = int(input("Enter the number of choice: "))

df = pd.DataFrame(columns=[0,1,2])
if totrow >= 1:
  df.loc[0] = [num,num,num]
if totrow >= 2:
  df.loc[1] = [num,num**2,num**3]
for i in range(2,totrow):
  df.loc[i] = [df.iloc[i-2, 0]+df.iloc[i-1, 0], df.iloc[i-1, 1]**2, df.iloc[i-1, 2]**3]

open("Pandas_exp.csv","w")
df.to_csv("Pandas_exp.csv",index = False)
print(df)
print()
with open("Pandas_exp.csv","r") as f:
  print(f.read())
```
