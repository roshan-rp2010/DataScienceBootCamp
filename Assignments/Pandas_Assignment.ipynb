Ch3_Q45_Data_File.xlsx

```python
path = "/content/Ch3_Q45_Data_File.xlsx"
df = pd.read_excel(path)
df.head(5)
```

```python
df_new = df.loc[:,:"Footage"]
df_new.head()
```

```python
print(df_new.Rent.mean().round(2))
print(df_new.Rent.std().round(2))
```

```python
print(df_new.Footage.mean().round(2))
print(df_new.Footage.std().round(2))
```

```python
print("For rent (Coefficient of Variation) :", (df_new.Rent.std() / df_new.Rent.mean()) * 100)
print("For footage (Coefficient of Variation) :", (df_new.Footage.std() / df_new.Footage.mean()) * 100)
```

```python
# So, Footage exhibits greater relative dispersion
```
