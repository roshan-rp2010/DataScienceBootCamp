1) Swap two numbers using 3rd variable

```python
def Swapa(a: int,b: int):
  print("\nBefore swap,\nFirst number is:   ", a)
  print("Second number is:  ", b)

  tmp = b
  b = a
  a = tmp

  print("\nAfter swap,\nFirst number is:   ", a)
  print("Second number is:  ", b)


a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
Swapa(a,b)
```

2) Swap two numbers without using 3rd variable

```python
def Swapb(a: int,b: int):
  print("\nBefore swap,\nFirst number is:   ", a)
  print("Second number is:  ", b)

  a, b = b, a

  print("\nAfter swap,\nFirst number is:   ", a)
  print("Second number is:  ", b)


a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
Swapb(a,b)
```

3) Write a program to find greatest common divisor (GCD) or highest common factor (HCF) of given two numbers

```python
def GCD(a: int,b: int):
  big = a if a < b else b
  gcd = 1
  for i in range(big,1,-1):
    if a%i == 0 and b%i == 0:
      gcd = i
      break

  print("The GCD is:",gcd)


a = int(input("Please enter first number : "))
b = int(input("Please enter second number: "))
GCD(a,b)
```

6) Write a program to print number names of digits in a number like

SAMPLE INPUT 231
SAMPLE OUTPUT TWO THREE ONE


```python
def spellNum(a: int):
  spell = ["ZERO ","ONE ","TWO ","THREE ","FOUR ","FIVE ","SIX ","SEVEN ","EIGHT ","NINE "]

  inalpha = [spell[int(i)] for i in str(a)]

  name = ""
  for i in inalpha:
    name += i

  print(name)


a = int(input("Enter your number: "))
spellNum(a)
```

7) Write a program to print Fibonacci series till n terms take n =50

```python
def Fib(n: int):
  ser = [0,1]
  for i in range(2,n,1):
    ser.append(ser[i-1] + ser[i-2])

  print(ser)


a = int(input("Total fibonacci numbers to be printed: "))
Fib(a)
```

9) Write a program to check if a number is Armstrong or not

```python
def Armst(n):
  length = len(str(n))
  tmp = 0
  for i in str(n):
    tmp += int(i)**length

  print(tmp == n)


a = int(input("Enter a number to check if it's armstrong: "))
Armst(a)
```

14) Write a Python program to convert temperatures to and from Celsius and Fahrenheit.

```python
def CtoF(d,u: str):
  cel = 0
  far = 0
  if u.lower() == "c":
    cel = d
    far = (d * (9/5)) + 32
  else:
    far = d
    cel = (d - 32) * (5 / 9)

  print(f"In degrees celsius: {cel}")
  print(f"In degrees fahrenheit: {round(far,2)}")


num = int(input("Enter the degrees as integer: "))
while True:
  unit = input("Enter the unit as \"c\" or \"f\" :").lower()
  if unit != "c" and unit != "f":
    print("Must enter \"c\" or \"f\" !")
    continue
  break
CtoF(num,unit)
```

15) Write a Python program that takes two digits m (row) and n (column) as input and generates a two-dimensional array. The element value in the i-th row and j-th column of the array should be i*j. else it will be 0

17) Print all possible combination for a string or a word

Sample input “ABC”

Sample output

BAC

CAB

CBA

ACB

BCA


```python
import itertools

def Perm(a: str):
  mylist = []
  mylist.extend(a)
  newlist = list(itertools.permutations(mylist))
  newlist = ["".join(i) for i in newlist]
  print(newlist)


a = input("Enter the string for it's permutation: ")
Perm(a)
```

18) Write a Python program to check if a triangle is equilateral, isosceles or scalene. Take 3 inputs for each side

Note :\
An equilateral triangle is a triangle in which all three sides are equal.\
A scalene triangle is a triangle that has three unequal sides.\
An isosceles triangle is a triangle with (at least) two equal sides.


```python
def CheckTri(a: int,b: int,c: int):
  if a == b and b == c:
    print("It is an equilateral Triangle.")
  elif a == b or a == c or a == c:
    print("It is an isosceles triangle.")
  else:
    print("It is a scalene triangle.")


a = int(input("Enter length of first side of the triangle:"))
b = int(input("Enter length of second side of the triangle:"))
c = int(input("Enter length of third side of the triangle:"))
CheckTri(a,b,c)
```
